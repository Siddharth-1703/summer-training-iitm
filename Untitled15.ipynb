```python
def even(num):
    if num % 2 == 0:
        return True
    else:
        return False
even(4)
```

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)
print(factorial(5)) 
```

```python
def palindrome(s):
    return s == s[::-1]
print(palindrome("madam")) 
print(palindrome("hello"))  
```

```python
def sum_all(*args):
    return sum(args)

# Example usag
print(sum_all(1, 2, 3))    
print(sum_all(5, 10, 15, 20))  
```
