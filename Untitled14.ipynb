```python
def family():
    try:
        family1male = int(input("Enter number of males in first family: "))
        family1female = int(input("Enter number of females in first family: "))
        family2male = int(input("Enter number of males in second family: "))
        family2female = int(input("Enter number of females in second family: "))
        if family1male > family1female:
            print("In first family, males are higher.")
        else:
            print("In first family, females are higher.")

        if family2male > family2female:
            print("In second family, males are higher.")
        else:
            print("In second family, females are higher.")
    except ValueError:
        print("Please enter valid integer numbers only.")
family()
```
