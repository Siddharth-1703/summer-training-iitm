```python
days=[1,3,5,7]
de=[38,42,36,39]
ch=[45,42,41,43]
ba=[28,26,24,22]
plt.plot(days,de,"mD--",label="Delhi")
plt.plot(days,ch,"c+--", label="Chennai")
plt.plot(days,ba,"g+--", label="Banglore")
plt.legend(loc="best")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Temperature in city")
```

```python
x=[10,20,30,40]
y=[25,45,65,80]
z=[20,30,40,50]
plt.plot(x,y,color="cyan", linewidth=5, ls="dotted", label="Line 1")
plt.plot(x,z, label="Line 2")
plt.legend()
plt.xlabel("Data 1")
plt.ylabel("Data 2")
plt.title("Line Chart")
```

```python
revenue=[120,90,80,55]
profit=[201,35,40,15]

company=["Google","Microsoft","facebook","Paytm"]
x=np.arange(len(company))
print(x)
plt.xticks(x,company)
plt.bar(x+0.2, revenue,width=0.4, label="Revenue")
plt.bar(x-0.2, profit, width=0.4, label="profit")
plt.legend()
plt.xlabel("Companies")
plt.ylabel("in Millions")
```
