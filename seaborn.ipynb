```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = {
    'total_bill': [10, 20, 30, 40, 50, 15, 25, 35, 45, 55],
    'tip': [1, 2, 3, 4, 5, 1.5, 2.5, 3.5, 4.5, 5.5],
    'sex': ['Male', 'Female']*5,
    'smoker': ['Yes', 'No']*5,
    'day': ['Thur', 'Fri', 'Sat', 'Sun', 'Thur', 'Fri', 'Sat', 'Sun', 'Thur', 'Fri'],
    'time': ['Lunch', 'Dinner']*5,
    'size': [1, 2, 3, 4, 1, 2, 3, 4, 2, 3]
}
df = pd.DataFrame(data)
sns.set(style='whitegrid')

#1. Barplot
sns.barplot(x='day', y="total_bill", data=df)  # Changed 'total bill' to 'total_bill' to match the column name in the DataFrame
plt.title("Barplot")
plt.show()

#2 Countplot
sns.countplot(x='day', data=df)
plt.title("Countplot")
plt.show()

#3. Boxplot
sns.boxplot(x='day', y='total_bill', data=df)
plt.title("Boxplot")
plt.show()

#4. Violinplot
sns.violinplot(x='day', y='total_bill', data=df)
plt.title("Violinplot")
plt.show()

#5. Stripplot
sns.stripplot(x='day', y='total_bill', data=df)
plt.title("Stripplot")
plt.show()

#6. Swarmplot
sns.swarmplot(x='day', y='total_bill', data=df)
plt.title("Swarmplot")
plt.show()

#7. Pointplot
sns.pointplot(x='day', y='total_bill', data=df)
plt.title("Pointplot")
plt.show()

# 8. Boxenplot
sns.boxenplot(x='day', y='total_bill', data=df)
plt.title("Boxenplot")
plt.show()
```
