```python
month = input("Enter the name of the month: ").strip().lower()
seasons = {
    'january': 'Winter',
    'february': 'Winter',
    'march': 'Spring',
    'april': 'Spring',
    'may': 'Spring',
    'june': 'Summer',
    'july': 'Summer',
    'august': 'Summer',
    'september': 'Autumn',
    'october': 'Autumn',
    'november': 'Autumn'
}
if month in seasons:
    print(f"The season in {month.capitalize()} is {seasons[month]}.")
else:
    print("Invalid month name. Please enter a valid month (e.g., January, February...).")
```
